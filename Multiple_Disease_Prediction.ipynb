```python
# step1: Import library
import pandas as pd                                                              # for data manipulation
import numpy as np                                                               # for numerical operations
import seaborn as sns                                                            # for advanced visualization (confusion matrix)
import matplotlib.pyplot as plt                                                  # for plotting graphs
```

```python
# Step 2: Read dataset
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
df.head()
```

```python
df.info(verbose=True)
```

```python
df.describe()
```

```python
# Step 3: Define y and X
df.columns
```

```python
# Preprocess Data
y = df['prognosis']
X = df.drop(['prognosis'], axis=1)
```

```python
df['prognosis'].value_counts()
```

```python
df['prognosis'].nunique()
```

```python
# Step 4: Split the Dataset
# We split the dataset into 2 parts: 75% for training, 25% for testing.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
df.shape, X_train.shape, X_test.shape
```

```python
# Step 5: Model selection
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# Step 6: Train the Model
model.fit(X_train,y_train)
```

```python
# Step 7: Predict the Model
y_pred = model.predict(X_test)
```

```python
# Step 8: Evaluate
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# Step 9: Confusion Matrix
from sklearn.metrics import confusion_matrix
plt.figure(figsize=(12, 8))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=False, cmap='coolwarm')
plt.title("Confusion Matrix (Multiple Diseases)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
